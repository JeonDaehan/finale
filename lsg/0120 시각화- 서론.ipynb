```python
ott=ott.drop("구분별(1)", axis=1)
ott=ott.drop("2022", axis=1)
```

```python
ott_age= ott.iloc[3:8]
ott_age
```

```python
ott_age=ott_age.set_index("구분별(2)")
```

```python
ott_age=ott_age.astype(float)
```

```python
ott_age=ott_age.rename(columns={"2022.1":"트위터","2022.2":"페이스북", "2022.3":"카카오스토리",
                      "2022.4":"인스타그램","2022.5":"네이버밴드"})
```

```python
colors=sns.color_palette("Set3")

ott_age.plot(kind="bar", title="주로 사용하는 SNS 순위", width=0.8,figsize=(8,8),color=colors)
plt.xlabel("나이대")
plt.ylabel("사용률")
```
