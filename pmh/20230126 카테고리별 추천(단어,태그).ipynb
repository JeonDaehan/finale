```python
# 불러오기
result = pd.read_csv("data/햄버거_R03.csv")
```

```python
result.info()
```

```python
result["Tags"] = result["Tags"].fillna("")
result.info()
```

```python
Tag = []
for i in result["Tags"]:
    ex = i[0:-1].split(", ")
    for j in ex:
        Tag.append(j)
Tag
```

```python
Tag_count = Counter(Tag)
Tag_count
```

```python
rank_text1 = Tag_count.most_common()
rank_text1
```

```python
rank_text1 = np.array(rank_text1)
rank_text1 = rank_text1[0:,0]
```

```python
rank_text1[0:50]
```

```python
Content = []
for i in result["Contents"]:
    ex = i[2:-2].split(" ")
    for j in ex:
        Content.append(j)
Content
```

```python
Content_count = Counter(Content)
Content_count
```

```python
rank_text2 = Content_count.most_common()
rank_text2
```

```python
rank_text2 = np.array(rank_text2)
rank_text2 = rank_text2[0:,0]
```

```python
rank_text2[0:50]
```
