```python
import re
from bs4 import BeautifulSoup

def get_user(driver):

    # 1. 현재 게시글 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 2. 게시물
    num_content = soup.select('span._ac2a')[0].text
    time.sleep(2)
    # 3. 팔로워
    follower = soup.select('span._ac2a')[1].text 
    time.sleep(2)

    # 4. 저장하기
    data = [num_content, follower]
    return data

get_user(driver)
```

```python
from tqdm.notebook import tqdm
import random

result = [ ]
for n in tqdm(range(1601,1747)):  
    try:
        name = result_user.iloc[n][0]
        url = searching(name)
        time.sleep(3)
        driver.get(url)
        time.sleep(5)
        
        #사람처럼 행동 random하게 스크롤 작동
        ran_cnt = random.randint(1,5)
        ran_length = random.choice(np.arange(100,200,100))
        for count in range(0,ran_cnt):
            driver.execute_script(f"window.scrollTo(0, {ran_length})")
            ran_length *= 2
            
        time.sleep(3)
        
        data = get_user(driver)
        result.append(data)
        time.sleep(5)
        
    except:
        data = ['0','0']
        result.append(data)
        
        #사람처럼 행동 random하게 스크롤 작동
        ran_cnt = random.randint(1,5)
        ran_length = random.choice(np.arange(100,200,100))
        for count in range(0,ran_cnt):
            driver.execute_script(f"window.scrollTo(0, {ran_length})")
            ran_length *= 2
        time.sleep(3)
```
