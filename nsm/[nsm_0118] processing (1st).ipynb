```python
import re
from bs4 import BeautifulSoup

def get_user(driver):

    # 1. 현재 게시글 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 2. 게시물
    num_content = soup.select('span._ac2a')[0].text

    # 3. 팔로워
    follower = soup.select('span._ac2a')[1].text 


    # 4. 저장하기
    data = [user_id, num_content, follower]
    return data

get_user(driver)
```

```python
### 수정 코드  (예외처리 추가+랜덤스크롤)
import random
from tqdm.notebook import tqdm

result = [ ]
for n in tqdm(range(len(tpk_user.index))):  
    
    #사람처럼 행동 random하게 스크롤 작동
    ran_cnt = random.randint(1,5)
    ran_length = random.choice(np.arange(100,200,100))
    for count in range(0,ran_cnt):
        driver.execute_script(f"window.scrollTo(0, {ran_length})")
        ran_length *= 2
        time.sleep(2)
        
    try:
        name = tpk_user.iloc[n][0]
        url = searching(name)
        
        driver.get(url)
        time.sleep(8)
    except:
        data = ['0','0']
        result.append(data)
        time.sleep(3)
    else:
        data = get_user(driver)    
        result.append(data)
        time.sleep(3)
```
