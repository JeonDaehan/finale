# 1차 전처리

1) 컬럼 이름 변경(공통)

2) contents 컬럼 특수 문자 뺀 문자열로 변경

3) 좋아요
    - 결측치(Nan) -> 0
    
    - 가장 먼저 눌러보세요 -> 0 
    
    - 여러명이 좋아합니다 -> 전체 좋아요수의 중간값 또는 평균값
    
    - '아이디'님이 좋아합니다 -> 1 

4) 좋아요 정규화(minmax scaler)     

5) 태그 수 파생 변수


## 떡볶이

```python
tpk1=pd.read_csv("../data/result_tpk_2000.csv", index_col = 0)
tpk1
```

```python
# 컬럼 이름 변경
tpk1.columns = ['ID','Contents','Tags', 'Likes']

#tpk1.index = list(range(1,501))
tpk1.head(5)
```

```python
import re
tpk1.iloc[0][1]

content = tpk1.iloc[0][1]
new_content = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "", str(content))

print(new_content)
```

```python
# 특수문자를 제외한 문자(한영), 숫자, 공백 가져오기

import re

for i in range(len(tpk1.index)):
    content = tpk1.iloc[i][1]
    new_content = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", "",  str(content))
    tpk1["Contents"][i] = new_content
    
tpk1.head()
```

```python

# 결측치 확인 및 제거

tpk1.info()

tpk1["Likes"].isnull().sum()
```

```python
tpk1["Likes"].fillna('0', inplace=True)
```

```python
# 가장 먼저 좋아요를 눌러보세요 처리 -> 0 처리

#tpk1[tpk1['Likes'] == '0'] # 2개
#tpk1[tpk1['Likes'] == '저 좋아요를 눌러보세'].nunique() # 22개

old = {'Likes' : {'저 좋아요를 눌러보세':'0'}}
tpk1 = tpk1.replace(old)

tpk1[tpk1['Likes'] == '0'].nunique() # 24개
tpk1[tpk1['Likes'] == '0']
```

```python
# "아이디"님 여러명이 좋아합니다 처리 -> 중간값 or 평균값 

tpk1[tpk1["Likes"].str.contains("여러")]
```

```python
# '여러' 포함한 행 드랍 -> tpk2

drop_row = tpk1[tpk1["Likes"].str.contains("여러")].index
drop_row

tpk2 = tpk1.drop(drop_row)
tpk2.info()

tpk2["Likes"]
```

```python
# 형식 바꾸기

# tpk2 = tpk2.astype({'Likes':'float'})
# tpk2.info()

# ValueError: could not convert string to float: 'dailyfood님이 좋아합니' -----> 에러 발생

old2 = {'Likes' : {'dailyfood님이 좋아합니':'1'}}
tpk2 = tpk2.replace(old2)

tpk2 = tpk2.astype({'Likes':'float'})
tpk2.info()


print('------------------------------------------------------')

mean_likes = round(np.mean(tpk2["Likes"]))
median_likes = np.median(tpk2["Likes"])


print(mean_likes)
print(median_likes)
```

```python
# 원 데이티셋의 중간값으로 변경  

idx = tpk1[tpk1["Likes"].str.contains("여러")].index
tpk1.loc[idx,'Likes'] = mean_likes

old3 = {'Likes' : {'dailyfood님이 좋아합니':'1'}}
tpk1 = tpk1.replace(old3)

tpk1 = tpk1.astype({'Likes':'float'})
tpk1.info()
```

```python
# 파생 변수 만들기

from sklearn.preprocessing import MinMaxScaler

MMscaler = MinMaxScaler()
tpk1["Scaled_Likes"]=MMscaler.fit_transform(tpk1[["Likes"]])

tpk1.head(10)
```

```python
# 파생 변수 만들기
```

```python
tpk1["Tags"] = tpk1["Tags"].str.strip("]['")
tpk1["Tags"] = tpk1["Tags"].str.replace("'", repl=r"", regex=True)
```

```python
tpk1["Tags"]
```

```python
tpk1["Tags"].str.split("#")[0]
```

```python
count=[]
for i in range(len(tpk1)):
    if len(tpk1["Tags"].str.split("#")[i]) >= 1:
        count.append(len(tpk1["Tags"].str.split("#")[i][1:]))
        
tpk1["Tag_count"] = count
```

```python
tpk1.sample(10)
```

```python
# 1차 전처리 저장
tpk1.to_csv("../data/tpk_processed_01.csv")
```

```python
import re
from bs4 import BeautifulSoup

def get_user(driver):

    # 1. 현재 게시글 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 아이디
    #id_ = soup.select('a > h2')[0].text
    id_ = soup.select('div > h2.x1lliihq')[0].text
    # 2. 게시물
    num_content = soup.select('span._ac2a')[0].text

    # 3. 팔로워
    follower = soup.select('span._ac2a')[1].text 


    # 4. 저장하기
    data = [id_, num_content, follower]
    return data

get_user(driver)
```
